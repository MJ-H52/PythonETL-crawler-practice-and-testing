```python
import openpyxl
import requests
from bs4 import BeautifulSoup

url = "https://www.104.com.tw/jobs/search/?ro=0&keyword=%E8%A6%96%E8%A6%BA%E5%B7%A5%E7%A8%8B%E5%B8%AB&expansionType=area,spec,com,job,wf,wktm&jobsource=2018indexpoc&langFlag=0&langStatus=0&recommendJob=1&hotJob=1&order=11"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

job_titles = []
company_names = []
job_contents = []
salaries = []

for job in soup.select(".js-job-item"):
    job_title = job.get("data-job-name")
    company_name = job.get("data-cust-name")
    job_content = job.select_one(".job-list-item__info").text.strip()
    salary = job.select_one(".b-tag--default").text.strip()

    if job_title:
        job_titles.append(job_title)
    
    if company_name:
        company_names.append(company_name)
    
    if job_content:
        job_contents.append(job_content)
    
    if salary:
        salaries.append(salary)

# 寫入 Excel 檔案
workbook = openpyxl.Workbook()
sheet = workbook.active
sheet.append(["工作名稱", "公司名稱", "工作內容", "薪資"])

for i in range(len(job_titles)):
    sheet.append([job_titles[i], company_names[i], job_contents[i], salaries[i]])

workbook.save("jobs.xlsx")

print("資料已寫入 jobs.xlsx 檔案")
```
